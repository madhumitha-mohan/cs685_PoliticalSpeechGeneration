```python
import pandas as pd
import numpy as np
import os
```

```python
base_dir = "/Users/mad/Desktop/UMass_CS/cs685/project/"

keywords_file = open(base_dir+"data/phrase_clusters/keywords.txt", "r")
keywords = keywords_file.readlines()
```

```python
topic_dict = {}
for line in keywords:

    [topic, phrase] = line.strip("\n").split("|")
    
    if topic=="topic":
        continue
    print("Topic: ", topic)
    print("Stemmed keyword: ", phrase)
    full_word = input("Enter full word:")
    
    if topic not in topic_dict.keys():
        topic_dict[topic] = []
    topic_dict[topic].append(full_word)
        
    
        
```

```python
for topic in topic_dict.keys():
    print(topic)
    print("\n")
    with open(base_dir + 'data/bow_topics/'+ topic + '.txt', 'w') as f:
        for word in topic_dict[topic]:
            print(word)
            f.write(word)
            f.write('\n')
```
